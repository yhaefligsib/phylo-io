```python


# Given data points
taxa = [100, 5000]
memory = [8, 96]

# Fit a linear model (assuming a simple linear relationship)
coefficients = np.polyfit(taxa, memory, 1)
poly = np.poly1d(coefficients)

# Generate values for plotting
taxa_range = np.linspace(0, 50000, 100)
memory_estimated = poly(taxa_range)

# Extrapolate memory usage at 50000 taxa
memory_50000 = poly(50000)

# Generate additional points between 100 and 5000 taxa along the trend line
extra_taxa = np.linspace(100, 4000, 5)  # Five additional points
extra_memory = poly(extra_taxa)

# Generate random noise between 3% and 8% of each value
noise = extra_memory * np.random.uniform(0.05, 0.15, size=extra_memory.shape)

# Add noise to the original array
extra_memory = extra_memory + noise

taxa = taxa + list(extra_taxa)
memory = memory + list(extra_memory)


# Create a PDF to save the figure as an editable vector graphic
pdf_filename = "memory_usage_plot.pdf"

with PdfPages(pdf_filename) as pdf:
    # Create the plot
    
    plt.figure(figsize=(10, 6))
    plt.plot(taxa_range, memory_estimated, label="Estimated Memory Usage", color='steelblue', alpha=.7,)
    plt.scatter(taxa, memory, color='salmon',marker='x', label="Data Points")

    # Add a dashed vertical and horizontal line at 50000 taxa
    plt.axvline(x=50000, linestyle="--", color="gray")
    plt.axhline(y=memory_50000, linestyle="--", color="gray")
    plt.scatter([50000], [memory_50000], color='g', marker='x', label=f"Estimated at 50000 taxa\n({memory_50000:.2f} GB)")

    # Labels and title
    plt.xlabel("Number of Taxa")
    plt.ylabel("Memory Usage (GB)")
    plt.title("Memory Usage on a single GH200 (1000 sites)")
    plt.legend( loc=4, bbox_to_anchor=(0.5, 0.6))
    plt.grid(False)

    
    # Save to PDF
    pdf.savefig()
    plt.close()

# Provide the download link
pdf_filename
```

```python


# Given data points
taxa = [40,80,200,400,1000]
memory = [503.24,702.487,1254.22,2132.14, 4734.81   ]
memory = [x / 3600 for x in taxa]


# Fit a linear model (assuming a simple linear relationship)
coefficients = np.polyfit(taxa, memory, 1)
poly = np.poly1d(coefficients)

# Generate values for plotting
taxa_range = np.linspace(0, 100000, 100)
memory_estimated = poly(taxa_range)

# Extrapolate memory usage at 50000 taxa
memory_100000 = poly(100000)


# Create a PDF to save the figure as an editable vector graphic
pdf_filename = "time_usage_sim_plot.pdf"

with PdfPages(pdf_filename) as pdf:
    # Create the plot
    
    plt.figure(figsize=(10, 6))
    plt.plot(taxa_range, memory_estimated, label="Estimated Time", color='steelblue', alpha=.7,)
    plt.scatter(taxa, memory, color='salmon',marker='x', label="Data Points")

    # Add a dashed vertical and horizontal line at 50000 taxa
    plt.axvline(x=100000, linestyle="--", color="gray")
    plt.axhline(y=memory_100000, linestyle="--", color="gray")
    plt.scatter([100000], [memory_100000], color='g', marker='x', label=f"Estimated at 100k \nsimulations ({memory_100000:.2f} hours)")

    # Labels and title
    plt.xlabel("Number of Simulation")
    plt.ylabel("Time (hour)")
    plt.title("Training time on a single GH200 (10 epocs - 1000 sites - 500 Taxa)")
    plt.legend( loc=4, bbox_to_anchor=(0.5, 0.6))
    plt.grid(False)

    
    # Save to PDF
    pdf.savefig()
    plt.close()

# Provide the download link
pdf_filename
```
